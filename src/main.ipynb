```python
# Figure size (width, height)

fig = plt.figure(figsize=(6, 6))

# Adjust the subplots

fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# For each of the 64 images

for i in range(64):

    # Initialize the subplots: add a subplot in the grid of 8 by 8, at the i+1-th position

    ax = fig.add_subplot(8, 8, i+1, xticks=[], yticks=[])

    # Display an image at the i-th position

    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')

    # Label the image with the target value

    ax.text(0, 7, str(digits.target[i]))

plt.show()



```

```python
print(f" Now that we know what we are working on let cluster the {(len(digits.images))} into group.")
```

```python
print(f"let find the k number of the cluster by using the elbow method")
inertias=[]
num_cluster = list(range(1,10))
for k in num_cluster:
    model = KMeans(n_clusters = k, random_state=42 )
    model.fit(digits.data)
    inertias.append(model.inertia_)
    # print(model.inertia_)

plt.plot(num_cluster, inertias,'-o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()


# Apres avoir avoir trouver le nombre de cluster je passe a l'entrainement du model sur le digits.data
k=4
model = KMeans(n_clusters =k , random_state=42)
model.fit(digits.data)

# affichage des centroides sous forme d'images --- dans le cours cetait sous forme de point
fig = plt.figure(figsize=(8,3))
fig.suptitle('Cluster center images',fontsize=20,fontweight='bold')

for i in list(range(1,10)):
    #Initialize subplot in a grid of 2x5, at i + 1th position
    ax = fig.add_subplot(2,5,i+1)

    # display images
    ax.imshow(model.cluster_centers_[i].reshape((8, 8)), cmap=plt.cm.binary)
plt.show()





```
